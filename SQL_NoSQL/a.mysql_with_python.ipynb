```python
import MySQLdb, pickle
import pandas as pd
```

```python
# pw = pickle.load(open('pw.p','rb'))
pw = '123456Cc'
```

##### Connect

```python
db = MySQLdb.connect(
    "127.0.0.1",
    "root",
    pw,
    "city_info",
    charset='utf8',
)
```

##### Select

```python
SQL_QUERY = """
    SELECT *
    FROM popular;
"""
```

```python
curs = db.cursor()
count = curs.execute(SQL_QUERY)
rows = curs.fetchall()
for row in rows:
    print(row)
# db.close()
```

##### insert, update, delete

```python
SQL_QUERY_2 = """
    INSERT INTO popular(rank, city_name, popular)
    VALUES(6, "gwanggju", 1502881)
"""
```

```python
curs = db.cursor()
count = curs.execute(SQL_QUERY_2)
db.commit()
# db.close()
```

##### database -> dataframe

```python
city_df = pd.read_sql(SQL_QUERY, db)
city_df
```

```python
import seaborn as sns
from sqlalchemy import create_engine
```

```python
tips = sns.load_dataset("tips")
tips.tail()
```

```python
engine = create_engine("mysql+mysqldb://root:" + pw + "@localhost/city_info")

tips.to_sql(name="tips", con=engine, if_exists='replace')
```
