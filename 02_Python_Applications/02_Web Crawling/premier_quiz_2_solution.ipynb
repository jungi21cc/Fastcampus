아래 페이지의 테이블을 크롤링하여 pandas 데이터 프레임으로 만들어 주세요.
- https://www.premierleague.com/tables

```python
from selenium import webdriver
import pandas as pd
```

```python
# make dataframe 
columns = ["Position", "Club", "Played", "Won", "Drawn", "Lost", "GF", "GA", "GD", "Points"]
df = pd.DataFrame(columns=columns)

# open window
driver = webdriver.Chrome()
driver.get("https://www.premierleague.com/tables")

# close layer
try:
    driver.find_element_by_css_selector("#advertClose").click()
except:
    print("no layer")

# select teams
table = driver.find_elements_by_css_selector("#mainContent .tableBodyContainer")[0]
teams = table.find_elements_by_css_selector("tr:not(.expandable)")

# crawling data
for team in teams:
    data = {
        "Position": team.find_element_by_css_selector("td.pos").text,
        "Club": team.find_element_by_css_selector("td.team .long").text,
        "Played": team.find_element_by_css_selector("td:nth-child(4)").text,
        "Won": team.find_element_by_css_selector("td:nth-child(5)").text,
        "Drawn": team.find_element_by_css_selector("td:nth-child(6)").text,
        "Lost": team.find_element_by_css_selector("td:nth-child(7)").text,
        "GF": team.find_element_by_css_selector("td:nth-child(8)").text,
        "GA": team.find_element_by_css_selector("td:nth-child(9)").text,
        "GD": team.find_element_by_css_selector("td:nth-child(10)").text,
        "Points": team.find_element_by_css_selector("td:nth-child(11)").text
    }
    df.loc[len(df)] = data

# close driver
driver.close()

# save csv
df.to_csv("premierleague.csv", encoding="utf-8")

# display dataframe
df
```
