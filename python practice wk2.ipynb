```python
data = "aabbbccqqddd"

encoded = ""

count = 1

for i in range(1, len(data)):   
    
    if data[i] == data[i - 1]:
        count += 1
    
    elif i == len(data) - 1:
        encoded += data[i] + str(count)
    
    else:
        encoded += data[i - 1] + str(count)
    
    

print(encoded)
```
