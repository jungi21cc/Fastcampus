### if, elif, else
- syntax

```
if condition_1:
    code_1
elif condition_2:
    code_2
else: 
    code_3
```
- condition이 True이면 해당 조건안쪽의 코드가 실행

````python
flag = True
if flag:
    print("if")
else:
    print("else")
````

````python
flag = False
if flag:
    print("if")
else:
    print("else")
````

````python
# 비교 연산자 사용
a, b = 10, 20
if a < b:
    print("a < b")
elif a == b:
    print("a == b")
else:
    print("a > b")
````

````python
# 논리 연산자 사용
a, b = 10, 20
if a < 20 and b < 10:
    print("code 1")
else:
    print("code 2")
````

````python
# 논리 연산자 사용
a, b = 10, 20
if a < 20 and not(b < 10):
    print("code 1")
else:
    print("code 2")
````

````python
# len(a) == 0 과 같이 사용하지 않는다.
a, b = [], [1,2]
if a:
    print("a")
if b:
    print("b")
````

### while
- syntax
```
while condition:
    code
```
- condition이 True이면 while 안쪽의 code가 실행 후 다시 condition을 확인하여 True이면 code를 실행을 반복
- break
- continue

````python
a = 3
while a > 0:
    print(a)
    a -= 1
````

````python
a = 3
while a > 0:
    print(a)
    if a == 2:
        break
    a -= 1
````

````python
a = [1,2,3,4,"q"]
idx = 0
while True:
    data = a[idx]
    idx += 1
    if data == 'q':
        break
    elif data % 2 == 0:
        continue
    print(data)
````

### for
- syntax
```
for value in list :
    code
```
- list에 있는 value를 순서대로 하나씩 꺼내서 code를 실행
- range
- zip
- enumerate

````python
ls = [0,1,4,7,9]
for value in ls:
    print(value)
````

````python
# range(end)
for value in range(5):
    print(value)
````

````python
# range(start:end)
for value in range(2,5):
    print(value)
````

````python
# range(start:end:step)
for value in range(1,6,2):
    print(value)
````

````python
for value in range(5,0,-1):
    print(value)
````

````python
for value in range(5,1,-1):
    print(value)
    
print()

for value2 in range(5,1, -2):
    print(value2)
````

````python
a = 0
for _ in range(5,1,1):
    a += 1
print(a)
````

````python
a = 0
for _ in range(5):
    a += 1
print(a)
````

##### zip
- 2개의 리스트를 key, value 형태로 묶어주는 함수

````python
# 2개의 리스트를 매핑시켜 딕셔너리 형태로 변경하세요
subjects = ["korean", "english", "math", "science"]
points = [100, 80, 90, 60]
````

````python
subject = ["korean", "English", "Math"]
points = [100, 80, 60,70]
convert_dic = {}
for idx in range(len(subject)):
    convert_dic[subject[idx]] = points[idx]
convert_dic
````

````python
convert_dict = {}
for idx in range(len(subjects)):
    convert_dict[subjects[idx]] = points[idx]
convert_dict
````

````python
converted_dict = {}
for subject, point in zip(subjects, points):
    print(subject, point)
    converted_dict[subject] = point
converted_dict
````

````python
tuple(zip(subjects, points)), dict(zip(subjects, points))
````

##### enumerate
- 리스트 형태의 데이터를 index와 value 값을 동시에 사용할수 있는 함수

````python
# list(enumerate(range(5,10)))
subjects = ["korean", "english", "math", "science"]
for idx, value in enumerate(subjects):
    print(idx, value)
````

### comprehention
- 반복문의 결과를 바로 리스트로 만들어 주는 방법
- 리스트에 1~5까지 숫자를 대입하는 방법

````python
# method 1
ls = []
ls = [1, 2, 3, 4, 5]
print(ls)
````

````python
# method 2
ls = []
ls.append(1)
ls.append(2)
ls.append(3)
ls.append(4)
ls.append(5)
print(ls)
````

````python
# method 3
ls = []
for num in range(1,6):
    ls.append(num)
print(ls)
````

````python
# method 4
ls = []
ls = [num for num in range(1,6)]
print(ls)
````

````python
# comprehention는 for문을 사용한는 것보다 속도가 빠릅니다.
````

````python
%%time
# compare excute time
ls = []
for num in range(1,10000):
    ls.append(num)
````

````python
%%time
# comprehention 사용
ls = []
ls = [num for num in range(1,10000)]
````

- names = ["kim test", "park python", "lee data", "jung science", "lee school"] (입력)
- names = ["kim test", "park python", "data", "jung science", "school"] (결과)

````python
# TODO
````

````python
names = ["kim test", "park python", "lee data", "KIM science", "lee school"]
last = []
for last_name in names:
    if last_name.split(" ")[0].lower() == "kim":
        print()
    else:
        last.append(last_name)

print(last)
    
````
