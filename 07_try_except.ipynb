### try & except
- try, except를 통해 에러를 컨트롤 한다.

```python
l = [1,2,3]
print(ls[3])
```

```python
try:
    print(ls[3])
except:
    print("except")
print("end!")
```

```python
try:
    print(l[3])
except IndexError as e:
    print(e)
```

```python
5/0
```

```python
try:
    5/0
except ZeroDivisionError as e:
    print(e)
```

```python
try:
    5/0
except Exception as e:
    print(e)
```

```python
try:
    l[3]
#     5/0
except IndexError as e:
    print(e)
except ZeroDivisionError as e:
    print(e)
```

```python
try:
#     l[3]
    5/0
except (IndexError, ZeroDivisionError) as e:
    print(e)
```

```python
# avoid error
try:
    l[3]
except (IndexError, ZeroDivisionError) as e:
    pass
```

### finally
- 예외 처리 코드를 실행후 항상 마지막으로 실행

```python
isError = False
try:
    5/0
except Exception as e:
    isError = True
finally:
    print("is error : {}".format(isError))
```

### Make Error
- Exception을 상속받아 새로운 에러를 만들수 있다.
- raise 를 통하여 에러를 발생시킬수 있다.

```python
class MakeError(Exception):
    
    def __str__(self):
        return "insert upper 10"

def input_number(num):
    if num >= 10:
        raise MakeError()
    print(num)
    
input_number(10)
```

```python
class MakeError(Exception):
    
    def __init__(self, msg):
        self.msg = msg

    def __str__(self):
        return self.msg

def input_number(num):
    if num >= 10:
        raise MakeError("insert upper 10")
    print(num)
    
try:
    input_number(5)
    input_number(11)
except MakeError as e:
    print(e)
```
